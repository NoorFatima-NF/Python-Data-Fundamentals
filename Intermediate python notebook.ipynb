```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(123)
```

```python
all_walks = []
for i in range(5):
    random_walk = [0]
    for x in range(100):
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)
```

```python
np_aw = np.array(all_walks)
```

```python
plt.plot(np_aw.T)  
plt.title("Random Walks (Original)")
plt.xlabel("Step")
plt.ylabel("Position")
plt.show()
```

```python
plt.clf()
```

```python
np_aw_t = np_aw.T
```

```python
plt.plot(np_aw_t)
plt.title("Random Walks (Transposed)")
plt.xlabel("Step")
plt.ylabel("Position")
plt.show()
```

```python
all_walks = []
for i in range(500):
    random_walk = [0]
    for x in range(100):
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        if np.random.rand() <= 0.001:
            step = 0

        random_walk.append(step)

    all_walks.append(random_walk)
```

```python
np_aw_t = np.transpose(np.array(all_walks))
ends = np_aw_t[-1, :]
```

```python
plt.hist(ends)
plt.show()
```
