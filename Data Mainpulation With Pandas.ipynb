GROUPED SUMMARY STATISTICS:

```python
import pandas as pd
sales = pd.DataFrame({
    "store": [1, 2, 3, 4, 5, 6],
    "type": ["A", "B", "C", "A", "B", "C"],
    "weekly_sales": [2000, 3000, 2500, 1500, 3500, 4000]
})
```

```python
sales_all = sales["weekly_sales"].sum()
sales_A = sales[sales["type"] == "A"]["weekly_sales"].sum()
sales_B = sales[sales["type"] == "B"]["weekly_sales"].sum()
sales_C = sales[sales["type"] == "C"]["weekly_sales"].sum()
sales_propn_by_type = [sales_A, sales_B, sales_C] / sales_all
print(sales_propn_by_type)
```

```python
sales_by_type = sales.groupby("type")["weekly_sales"].sum()
sales_propn_by_type = sales_by_type / sales_by_type.sum()
print(sales_propn_by_type)
```

```python
import pandas as pd
sales = pd.DataFrame({
    "store": [1, 2, 3, 4, 5, 6],
    "type": ["A", "B", "C", "A", "B", "C"],
    "weekly_sales": [2000, 3000, 2500, 1500, 3500, 4000],
    "unemployment": [5.0, 6.2, 4.8, 5.5, 6.0, 4.9],
    "fuel_price_usd_per_l": [1.2, 1.3, 1.1, 1.25, 1.35, 1.15]
})
```

```python
sales_stats = sales.groupby("type")["weekly_sales"].agg(["min", "max", "mean", "median"])
print("Weekly Sales Stats by Type:")
print(sales_stats)
```

```python
unemp_fuel_stats = sales.groupby("type")[["unemployment","fuel_price_usd_per_l"]].agg(["min", "max", "mean", "median"])
print("\nUnemployment and Fuel Price Stats by Type:")
print(unemp_fuel_stats)
```

WORKING WITH PIVOT TABLE:

```python
import pandas as pd
temperatures = pd.DataFrame({
    "country": ["India","India","India","Iraq","Iraq","Iraq"],
    "city": ["Hyderabad","Hyderabad","Delhi","Baghdad","Baghdad","Basra"],
    "date": pd.to_datetime([
        "2010-01-01","2011-01-01","2010-06-01",
        "2010-03-01","2011-04-01","2011-07-01"
    ]),
    "avg_temp_c": [30,32,35,28,31,40]
})
```

```python
temperatures["year"] = temperatures["date"].dt.year
```

```python
temp_by_country_city_vs_year = temperatures.pivot_table(
    values="avg_temp_c",
    index=["country","city"],
    columns="year"
)
```

```python
print(temp_by_country_city_vs_year)
```

```python
import pandas as pd
temperatures = pd.DataFrame({
    "country": ["Egypt","Egypt","India","India","India","Pakistan"],
    "city": ["Cairo","Alexandria","Delhi","Hyderabad","Mumbai","Karachi"],
    "year": [2005, 2008, 2005, 2007, 2010, 2009],
    "avg_temp_c": [22, 24, 28, 30, 32, 29]
})
```

```python
temp_by_country_city_vs_year = temperatures.pivot_table(
    values="avg_temp_c",
    index=["country","city"],
    columns="year"
)
```

```python
print(temp_by_country_city_vs_year.loc["Egypt":"India"])
print(temp_by_country_city_vs_year.loc[("Egypt","Cairo"):("India","Delhi")])
print(temp_by_country_city_vs_year.loc[("Egypt","Cairo"):("India","Delhi"), 2005:2010])
```

```python
mean_temp_by_year = temp_by_country_city_vs_year.mean()
print(mean_temp_by_year[mean_temp_by_year == mean_temp_by_year.max()])
```

```python
mean_temp_by_city = temp_by_country_city_vs_year.mean(axis=1)
print(mean_temp_by_city[mean_temp_by_city == mean_temp_by_city.min()])
```

READING AND WRITING CSVS:

```python
import pandas as pd
```

```python
airline_bumping = pd.DataFrame({
    "airline": ["Delta", "United", "Southwest", "JetBlue"],
    "bumps": [150, 120, 95, 80],
    "flights": [50000, 48000, 47000, 45000]
})
```

```python
import pandas as pd

airline_totals = pd.DataFrame({
    "airline": ["Delta", "United", "Southwest", "JetBlue"],
    "nb_bumped": [310, 250, 195, 150],
    "total_passengers": [102000, 97000, 94500, 85000]
})
```

```python
airline_totals.set_index("airline", inplace=True)
airline_totals["bumps_per_10k"] = airline_totals["nb_bumped"] / airline_totals["total_passengers"] * 10000
airline_totals_sorted = airline_totals.sort_values(by="bumps_per_10k", ascending=False)
```

```python
print(airline_totals_sorted)
```
